```python
from PIL import ImageEnhance
```

```python
enhancer = ImageEnhance.Brightness(img)
```

```python
display(enhancer.enhance(2))
```

```python
enhancer1 = ImageEnhance.Contrast(img)
```

```python
display(enhancer1.enhance(2))
```
